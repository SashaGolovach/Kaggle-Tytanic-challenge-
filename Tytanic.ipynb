## Customizing data set
* Normalized values
* United SibSp and Parch into 1 family
* Removed uneccessary columns

```python
%config IPCompleter.greedy=True
import pandas as pd
from matplotlib import pyplot as plt


df = pd.DataFrame(pd.read_csv('train.csv'))
df = df.drop(columns = ['Cabin', 'Embarked', 'Ticket', 'Name', 'PassengerId'])
df = df.dropna(axis = 0)
df['Sex'] = (df['Sex'] == 'male').astype(int)
df['Age'] /= 80 / 3
df['Fare'] /= 513 / 3
df['Parch'] /= 2
df['SibSp'] /= 5 / 3
df
```

```python
a = df.loc[df['Survived'] == 1, 'Fare']
plt.hist(a, bins=50)
plt.show()
```

```python
df.to_csv('data.csv')
```
